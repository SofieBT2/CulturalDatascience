## Music sales challenge

Write a script that:

1. Combines the tables of best-selling physical singles and best-selling digital singles on the Wikipedia page "List_of_best-selling_singles"
2. Adds a column which marks whether each row is from the list of physical singles or digital singles
3. Outputs the artist and single name for the year you were born. If there is no entry for that year, take the closest year after you were born.
4. Outputs the artist and single name for the year you were 15 years old.

```python
# Starter code...

#%pip install lxml
import pandas as pd

rawdata = pd.read_html("https://en.wikipedia.org/wiki/List_of_best-selling_singles")
```

```python
# Based on a quick look at Wikipedia, let's assume:
# Table 0 is the best-selling physical singles
# Table 3 is the best-selling digital singles

# Extract the relevant tables
physical_singles = rawdata[0]
digital_singles = rawdata[3]

# Combine the two tables using pd.concat
combined_singles = pd.concat([physical_singles, digital_singles])

# Print the combined table
combined_singles
#type(combined_singles)
```

```python
# Add a column which marks whether each row is from the list of physical singles or digital singles
Single_type= []
for x, row in combined_singles.iterrows():
    if row['Single'] in physical_singles['Single'].values:
        Type = 'Physical'
    else:
        Type = 'Digital'
    Single_type.append(Type)

Single_type
```

```python
combined_singles['Type'] = Single_type
combined_singles
```

```python
# Another way to do the previous task that alligns a bit more with the text. Although, I am still not satisfied with the fact that I have to know the bin sizes to manually label them
combined_singles['Type'] = pd.cut(x = combined_singles['Released'], 
                       bins = [1900,2000,2024], labels = ['physical', 'digital'])
combined_singles
```

```python
# Outputs the artist and single name for the year you were born. If there is no entry for that year, take the closest year after you were born.
# Sorting the dataframe by Released year
df = combined_singles.sort_values(by=['Released'])
df

# Check if the year 1999 is present
if 1999 in df['Released'].values:
    # If 1999 is found, filter by the exact year
    result = df[df['Released'] == 1999]
else:
    # If 1999 is not found, find the closest year after 1999
    result = df[df['Released'] > 1999].head(1)

# Print the artist and single for the matching year or closest year after
    print("Artist:", result['Artist'].values[0])
    print("Single:", result['Single'].values[0])
    print("Released:", result['Released'].values[0])
```

```python
# Outputs the artist and single name for the year you were 15 years old.
# Check if the year 2014 is present
if 2014 in df['Released'].values:
    # If 2014 is found, filter by the exact year
    result = df[df['Released'] == 2014]
else:
    # If 2014 is not found, find the closest year after 2014
    result = df[df['Released'] > 2014].head(1)

# Print the artist and single for the matching year or closest year after
    print("Artist:", result['Artist'].values[0])
    print("Single:", result['Single'].values[0])
    print("Released:", result['Released'].values[0])
```

## Space challenge

1. Make a single dataframe that combines the space missions from the 1950's to the 2020's
2. Write a script that returns the year with the most launches
3. Write a script that returns the most common month for launches
4. Write a script that ranks the months from most launches to fewest launches


```python
# Starter code...

#%pip install lxml
import pandas as pd

rawdata = pd.read_html("https://en.wikipedia.org/wiki/Timeline_of_Solar_System_exploration")
```

```python
# Make a single dataframe that combines the space missions from the 1950's to the 2020's
# Extract the relevant tables
Old = rawdata[0]
New = rawdata[7]

# Combine the two tables using pd.concat
Missions = pd.concat([Old, New])

# Print the combined table
Missions
```

```python
#Write a script that returns the year with the most launches
#Missions['Launch date']

# Convert 'Launch date' to datetime (ensure the column is in a proper date format)
Missions['Launch date'] = pd.to_datetime(Missions['Launch date'])

# Extract the year from the 'Launch date' column since that is part of the .to_datetime function
Missions['Year'] = Missions['Launch date'].dt.year

# Count the number of launches per year
launches_per_year = Missions['Year'].value_counts()

# Find the year with the most launches
most_launches_year = launches_per_year.idxmax()
most_launches_count = launches_per_year.max()

# Print the result
print(f"The year with the most launches is {most_launches_year} with {most_launches_count} launches.")
```

```python
#Write a script that returns the most common month for launches
# Extract the month from the 'Launch date' column since that is part of the .to_datetime function
Missions['Month'] = Missions['Launch date'].dt.month

# Count the number of launches per month
launches_per_month = Missions['Month'].value_counts()

# Find the most common month
most_common_month = launches_per_month.idxmax()
most_common_month_count = launches_per_month.max()

# Print the result
print(f"The most common month for launches is month number {most_common_month} with {most_common_month_count} launches.")
```

```python
# Write a script that ranks the months from most launches to fewest launches

# Count the number of launches per month and sort in descending order
launches_per_month = Missions['Month'].value_counts().sort_values(ascending=False)

# Print the ranked months and their corresponding number of launches
print("Months numbers ranked from most launches to fewest launches:")
launches_per_month
```

## Stroop challenge

Every year between 2015 and 2021, the students in my Language, Cognition, and the Brain course participated in a version of the Stroop task. Using a stopwatch (ok, using their phones), they recorded how fast they could say a list of things (either reading or naming colors or color words). The column names mean "Reading with No Interference", "Naming with Interference", "Naming with No Interference", and "Reading with Interference". The times are in seconds.

### Stroop challenge 1: 
Transform these data from wide format to long format, so that the result is a dataframe with
- 1 column named "Participant_id" with a unique number for each participant (you can use the row indices)
- 1 column named "Year" with the year data
- 1 column named "Task" that shows which task they were doing
- 1 column named "RT" that shows their response time

```python
# Starter code 1...

df = pd.read_csv("https://raw.githubusercontent.com/ethanweed/Stroop/master/Stroop-raw-over-the-years.csv")
df.head()
```

```python
# Starter code 2...

# Make a new column using the dataframe indices as particpant numbers

df.index.name = 'Participant_id'
df = df.reset_index()

#df.reset_index(inplace = True)
# NOTE: This line does exactly the same thing as the line above:
# it replaces the original df with a new df with the updated index. That's what
# "inplace = True" means. Or, you can just assign the dataframe with the updated index
# to a new dataframe with the same name as the old dataframe, which is what I did above.
# The end result is the same.

df
```

```python
df_long = pd.wide_to_long(df, 
                          stubnames = ['Reading', 'Naming'], 
                          i=['Participant_id', 'Year'], 
                          j='Task',
                          sep = '_', 
                          suffix = '.+')
df_long.head(10)
```
